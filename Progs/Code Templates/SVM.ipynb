## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
X, y = dataset.data, dataset.target
```

```python
for i in range(4) :
    print("Corr ", (i+1), " : " , np.corrcoef(X[:, i], y)[1][0])
    plt.scatter(X[:, i], y)
    plt.title("Relation between \nOutput and Feature "+ dataset.feature_names[i])
    plt.xlabel("Feature " + str(i+1))
    plt.ylabel("Output")
    plt.show()
    
```

```python
from sklearn.preprocessing import StandardScaler
stdScaler = StandardScaler()
X = stdScaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.25, stratify=y)
```

# SVM

```python
from sklearn.svm import SVC
svc_model = SVC(kernel = 'rbf', random_state = 42)
svc_model = svc_model.fit(X_train, y_train)

print("Training Score using SVM   : ", svc_model.score(X_train, y_train))
print("Validation Score using SVM : ", svc_model.score(X_val, y_val))
```
